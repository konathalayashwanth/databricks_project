```python
spark.sql("use catalog retail_chain_catalog")

spark.sql("create  schema if not exists retail_chain_catalog.bronze_schema")


def load_to_bronze(table_name, source_load, checkpoint_read, checkpoint_write):
    path = f"{source_load}"
    checkpoint_reads = f"{checkpoint_read}"
    checkpoint_writes = f"{checkpoint_write}"
    
    return (spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("cloudFiles.schemaLocation", checkpoint_reads)
        .load(path)
        .writeStream
        .format("delta")
        .option("checkpointLocation", checkpoint_writes)
        .trigger(availableNow=True)
        .table(f"retail_chain_catalog.bronze_schema.{table_name}"))

# customers
load_to_bronze("customers", 
               "/Volumes/retail_chain_catalog/default/ingestion_vol/customers/",
                "/Volumes/retail_chain_catalog/default/checkpoint_vol/customers/customer_read_checkpoint/",
                 "/Volumes/retail_chain_catalog/default/checkpoint_vol/customers/customers_write_checkpoint/"
                 )
#customertypes
load_to_bronze("customer_types",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/customer_types/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/customer_types/customer_types_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/customer_types/customer_types_write_checkpoint/")
#accounts
load_to_bronze("accounts",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/accounts/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/accounts/accounts_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/accounts/accounts_write_checkpoint/")
#accounttypes
load_to_bronze("account_types",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/account_types/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/account_types/accountypes_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/account_types/account_types_write_checkpoint/")

#account_statuses
load_to_bronze("account_statuses",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/account_statuses/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/account_statuses/accountstatuses_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/account_statuses/account_statuses_write_checkpoint/")

#branches
load_to_bronze("branches",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/branches/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/branches/branches_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/branches/branches_write_checkpoint/")

#addresses
load_to_bronze("addresses",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/addreses/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/addresses/addresses_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/addresses/addresses_write_checkpoint/")

#transactions
load_to_bronze("transactions",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/transactions/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/transactions/transactions_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/transactions/transactions_write_checkpoint/")

#transaction_types
load_to_bronze("transaction_types",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/transation_types/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/transaction_types/transaction_types_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/transaction_types/transaction_types_write_checkpoint/")
#loans
load_to_bronze("loans",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/loans/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/loans/loans_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/loans/loans_write_checkpoint/")

#loan_statuses
load_to_bronze("loan_statuses",
               "/Volumes/retail_chain_catalog/default/ingestion_vol/loan_statuses/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/loan_statuses/loan_statuses_read_checkpoint/",
               "/Volumes/retail_chain_catalog/default/checkpoint_vol/loan_statuses/loan_statuses_write_checkpoint/")

```
