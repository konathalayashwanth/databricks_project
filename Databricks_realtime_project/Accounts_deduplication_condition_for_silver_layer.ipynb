```python
%sql
CREATE OR REPLACE TABLE  retail_chain_catalog.silver_schema.silver_accounts_final_dedupe
USING DELTA
AS
SELECT 
    AccountID,
    CustomerID,
    AccountTypeID,
    AccountStatusID,
    Balance,
    OpeningDate,
    _updated_at
FROM 
(
SELECT
*,
row_number() OVER(PARTITION BY AccountID ORDER BY _updated_at asc) as rn
FROM retail_chain_catalog.silver_schema.silver_accounts_cleaned
) As k
WHERE rn = 1;
```
