```python
%sql
CREATE OR REPLACE TABLE  retail_chain_catalog.silver_schema.silver_loans_final_dedupe
USING DELTA
AS
SELECT *
FROM
(
SELECT *,
    row_number() over(partition by LoanID  order by _updated_at asc) as rn
    FROM retail_chain_catalog.silver_schema.silver_loans_cleaned
) as k
WHERE rn = 1;
```
